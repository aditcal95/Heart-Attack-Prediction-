```python
#Header File

import json
import pandas as pd
import requests
import numpy as np
import plotly as py
import plotly.graph_objs as go
from sklearn import model_selection
from sklearn.cross_validation import KFold
import matplotlib.pyplot as plt
```

```python
#importing the Machine Learning Libraries and performing a comparative study between them
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
#Gathering Data:

File = "heartattack.csv"
Data = pd.read_csv(File)
Data
```

```python
#Summarizing Data to detect the NaN data

Columns = ['age','sex','cp','trestbps','chol','fbs', 'restecg','thalach','exang','oldpeak','slope','ca','thal','heartpred']

for index in Columns:
    print(index)
    print(Data[index].unique())
```

```python
#Data Filtering

#1.Replacing the HeartPred values to 1 as this code deals only with prediction of Attack and not the Catagory of attack

Data['heartpred'] = np.where(Data['heartpred'] == 2, 1, Data['heartpred'])
Data['heartpred'] = np.where(Data['heartpred'] == 3, 1, Data['heartpred'])
Data['heartpred'] = np.where(Data['heartpred'] == 4, 1, Data['heartpred'])

#2. Replacing all the NaN values with 0s:
Data["thal"] = Data["thal"].fillna(0)
Data["ca"] = Data["ca"].fillna(0)

#Result Post Filteration
Columns = ['ca','thal','heartpred']

for index in Columns:
    print(index)
    print(Data[index].unique())
    
Data.to_csv('Heart2.csv')
```

```python
# Preparing Data For Machine Learning 303x14

# Data Processing:
Arr = Data.values
DataSize = len(Data['heartpred'])
BatchSize = 10
X = Arr[:,0:13] #All the attributes
Y = Arr[:,13] #Target Data
Fold = KFold(DataSize, BatchSize, random_state=7)
```

```python
# Algorithm Implementation and Comparison

Algorithm = []
Algorithm.append(('LR', LogisticRegression()))
Algorithm.append(('KNN', KNeighborsClassifier()))
Algorithm.append(('DTC', DecisionTreeClassifier()))
Algorithm.append(('SVC', SVC()))
Algorithm.append(('RFC', RandomForestClassifier()))

Names = []
Accuracy = []
for name, model in Algorithm:
    Results=[]
    i=0
    count=0
    for Tr, Pred in Fold:
        X_Tr = X[Tr,:] #input for training 
        Y_Tr = Y[Tr] #output for training
        X_inPred = X[Pred,:] #input for prediction
        Y_actual = Y[Pred] #Actual output 
        model.fit(X_Tr, Y_Tr) #Fitting the algorithm into the training data
        result = model.predict(X_inPred) 
        Results.append(result)
    Results = np.concatenate(Results, axis=0)             
    for each in Data["heartpred"]: #to compute the accuracy of the algorithm that is being used
        if each==Results[i]:
            count+=1
        i+=1
    accuracy=count/i
    print("Name=")
    print(name)
    print("Accuracy = ")
    print(accuracy*100)
    Names.append(name)
    Accuracy.append(accuracy)
```

```python
#Cross Validating the algorithms for different batch sizes:

def batchsize(a):
    Fold = KFold(DataSize, a, random_state=7)
    # Algorithm Implementation and Comparison
    Algorithm = []
    Algorithm.append(('LR', LogisticRegression()))
    Algorithm.append(('KNN', KNeighborsClassifier()))
    Algorithm.append(('CART', DecisionTreeClassifier()))
    Algorithm.append(('SVM', SVC()))
    Algorithm.append(('RFC', RandomForestClassifier()))

    Names = []
    Accuracy = []
    for name, model in Algorithm:
        Results=[]
        i=0
        count=0
        for Tr, Pred in Fold:
            X_Tr = X[Tr,:]
            Y_Tr = Y[Tr]
            X_inPred = X[Pred,:]
            Y_actual = Y[Pred]
            model.fit(X_Tr, Y_Tr)
            result = model.predict(X_inPred)
            Results.append(result)
        Results = np.concatenate(Results, axis=0)
        for each in Data["heartpred"]:
            if each==Results[i]:
                count+=1
            i+=1
        accuracy=count/i
        Names.append(name)
        Accuracy.append(accuracy)
    return[Names,Accuracy]


[x1,y1] = batchsize(3)
[x2,y2] = batchsize(5)
[x3,y3] = batchsize(10)

print("Batch Size = 3")
print(x1)
print(y1)
print("Batch Size = 5")
print(x2)
print(y2)
print("Batch Size = 10")
print(x3)
print(y3)

py.offline.init_notebook_mode(connected=True)

trace1 = go.Bar(
    x=x1,
    y=y1,
    name='Batch Size = 3'
)
trace2 = go.Bar(
    x=x2,
    y=y2,
    name='Batch Size = 5'
)
trace3 = go.Bar(
    x=x3,
    y=y3,
    name='Batch Size = 10'
)
plt2 = [trace1, trace2, trace3]
layout = go.Layout(
    title='Cross Validation',
    xaxis=dict(
        title='Machine Learning Algorithm',
        titlefont=dict(
            family='Arial, serif',
            size=18,
            color='black'
        )
    ),
        yaxis=dict(
        title='Accuracy',
        titlefont=dict(
            family='Arial, serif',
            size=18,
            color='black'
        )
    )
)

fig2 = go.Figure(data=plt2, layout=layout)

py.offline.iplot(fig2, filename='basic-bar')
```
